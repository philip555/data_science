## Score

```python
train.columns
```

```python
feature_names = ['X', 'Y', 'Dates-year', 'Dates-month',
       'Dates-day', 'Dates-hour', 'Dates-minute', 'PdDistrict_BAYVIEW', 'PdDistrict_CENTRAL',
       'PdDistrict_INGLESIDE', 'PdDistrict_MISSION',
       'PdDistrict_NORTHERN', 'PdDistrict_PARK',
       'PdDistrict_RICHMOND', 'PdDistrict_SOUTHERN',
       'PdDistrict_TARAVAL', 'PdDistrict_TENDERLOIN',
       'Dayofweek_Friday', 'Dayofweek_Monday', 'Dayofweek_Saturday',
       'Dayofweek_Sunday', 'Dayofweek_Thursday', 'Dayofweek_Tuesday',
       'Dayofweek_Wednesday']
feature_names
```

```python
X_train = train[feature_names]

print(X_train.shape)
X_train.head(1)
```

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head(1)
```

```python
label_name = "Category"

y_train = train[label_name]
print(y_train.shape)
y_train.head(1)
```

## XGBoost

```python
from sklearn.model_selection import cross_val_score
import xgboost as xgb

seed = 33 

model = xgb.XGBClassifier(n_estimators=15,
                         seed=seed,
                         nthread=-1)
model
```

```python
%time score = cross_val_score(model, X_train, y_train, cv=5, scoring="neg_log_loss").mean()

print("Score = {0:.5f}".format(score))
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict_proba(X_test)

print(predictions.shape)
predictions[0:1]
```

```python
submission = pd.read_csv("sampleSubmission.csv", index_col="Id")

for i, column in enumerate(model.classes_):
    submission[column] = predictions[:, i]
    
print(submission.shape)
submission.head()
```

```python
submission.to_csv("xbg_feed_onehot.csv")
```
