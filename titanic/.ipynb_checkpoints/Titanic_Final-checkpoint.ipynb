```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
```

# Load Dataset

### train data

```python
train = pd.read_csv('train.csv', index_col="PassengerId")
print(train.shape)
train.head()
```

### test data

```python
test = pd.read_csv('test.csv', index_col="PassengerId")
print(test.shape)
test.head()
```

## Preprocessing

### sex encoding

```python
train.loc[train["Sex"] == 'male', "Sex_encode"] = 0
train.loc[train["Sex"] == 'female', "Sex_encode"] = 1

print(train.shape)
train[["Sex", "Sex_encode"]].head()
```

```python
test.loc[test["Sex"] == 'male', "Sex_encode"] = 0
test.loc[test["Sex"] == 'female', "Sex_encode"] = 1

print(test.shape)
test[["Sex", "Sex_encode"]].head()
```

### fill in missing "Fare"

```python
mean_fare = train["Fare"].mean()  # train data에서 평균값을 도출
test.loc[pd.isnull(test["Fare"]), "Fare"] = mean_fare  # 평균 운임요금 대입을 통해, 빈 데이터 제거
test[pd.isnull(test["Fare"])]
```

## Encode Embarked

```python
# 선착장 encoding

train_embarked = pd.get_dummies(train["Embarked"], prefix="Embarked")
print(train_embarked.shape)
train_embarked.head()
```

```python
train = pd.concat([train, train_embarked], axis=1)

print(train.shape)
train[["Embarked", "Embarked_C", "Embarked_Q", "Embarked_S"]]
```

```python
pd.get_dummies(test["Embarked"])
```

```python
test_embarked = pd.get_dummies(test["Embarked"], prefix="Embarked")
print(test_embarked.shape)
test_embarked.head()
```

```python
# 
test = pd.concat([test, test_embarked], axis=1)

print(test.shape)
test[["Embarked", "Embarked_C", "Embarked_Q", "Embarked_S"]]
```

```python
lable_name = "Survived"

y_train = train[lable_name]
print(y_train.shape)
y_train.head()
```
