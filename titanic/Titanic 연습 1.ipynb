```python
import pandas as pd
```

# Load Dataset

## train data

```python
train = pd.read_csv('train.csv', index_col="PassengerId")
print(train.shape)
train.head()
```

## teat data

```python
test = pd.read_csv('test.csv', index_col="PassengerId")
print(test.shape)
test.head()
```

# Preprocessing

## Sex encode

```python
# Sex 컬럼의 male = 0, female = 1로 encode

train.loc[train["Sex"] == 'male', "Sex_encode"] = 0 
train.loc[train["Sex"] == 'female', "Sex_encode"] = 1
```

```python
train[["Sex", "Sex_encode"]].head()
```

```python
test.loc[test["Sex"] == 'male', "Sex_encode"] = 0
test.loc[test["Sex"] == 'female', "Sex_encode"] = 1
test[["Sex", "Sex_encode"]].head()

# test.head()
```

## Filling nan value of "Fare" of test

```python
# .isnull() method

test[pd.isnull(test["Fare"])]
```

```python
### train dataframe에서 "Fare"의 평균값 내기

mean_fare = train["Fare"].mean()
test.loc[pd.isnull(test["Fare"]), "Fare"] = mean_fare
test.loc[1044]
```

## Embarked encode

```python
# .get_dummies method

train_embarked = pd.get_dummies(train["Embarked"], prefix="Embarked")
print(train_embarked.shape)
train_embarked.head()
```

```python
# .comcat 함수를 이용해 train_embarked 테이블 붙이기

train = pd.concat([train, train_embarked], axis=1)
print(train.shape)
train.head()
```

### test의 Embarked 컬럼 encode

```python
test_embarked = pd.get_dummies(test["Embarked"], prefix="Embarked")
print(test_embarked.shape)
test_embarked.head()
```

```python
# .concat([리스트])
test = pd.concat([test, test_embarked], axis=1)
print(test.shape)
test.head()
```

## 의사 결정에 도움을 주는 X축 만들기

```python
feature_names = [
    "Pclass", "Sex_encode", "Fare", "Embarked_C", "Embarked_Q", "Embarked_S" 
]
X_train = train[feature_names]
print(X_train.shape)
X_train.head()
```

## 찾아야 하는 값인 y_label

```python
label = "Survived"

y_train = train[label]
```

```python
feature_names = [
    "Pclass", "Sex_encode", "Fare", "Embarked_C", "Embarked_Q", "Embarked_S" 
]
X_test = test[feature_names]
```

```python
print(X_test.shape)
X_test.head()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth = 5)
model
```

```python
model.fit(X_train, y_train)
```

```python
%load_ext gvmagic
from sklearn.tree import export_graphviz

dot_tree = export_graphviz(model, feature_names=feature_names, out_file=None)

%dotstr dot_tree
```

```python
predictions = model.predict(X_test)
print(predictions.shape)
predictions
```
