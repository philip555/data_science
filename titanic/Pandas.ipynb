## One Hot Encoding

```python
# .get_dummies를 통해서 탑승장 encoding하기
# prefix를 통해, 컬럼명을 명확하게 해줌

embarked = pd.get_dummies(data["Embarked"], prefix="Embarked")
```

```python
embarked.head()
```

```python
# concat means concatenate ( 붙이다 )

data_embarked = pd.concat([data, embarked], axis=1)
```

```python
data_embarked.head()
```

# 함수를 이용해서, 문자열의 특정구간만 가져오기

```python
def get_ticket_id(ticket):
    return ticket[0:3]

data["Ticket"].apply(get_ticket_id)
```
